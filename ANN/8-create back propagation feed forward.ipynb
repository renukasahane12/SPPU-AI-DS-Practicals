```python
import numpy as np

np.random.seed(42)

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of sigmoid
def sigmoid_derivative(x):
    return x * (1 - x)

# Input dataset (XOR)
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

# Output dataset
y = np.array([[0],
              [1],
              [1],
              [0]])

# Hyperparameters
learning_rate = 0.1
num_epochs = 100000

# Initialize weights
hidden_weights = 2 * np.random.random((2, 2)) - 1
hidden_bias = np.zeros((1, 2))
output_weights = 2 * np.random.random((2, 1)) - 1
output_bias = np.zeros((1, 1))

# Training
for i in range(num_epochs):
    # Forward propagation
    hidden_layer_input = np.dot(X, hidden_weights) + hidden_bias
    hidden_layer = sigmoid(hidden_layer_input)
    output_layer_input = np.dot(hidden_layer, output_weights) + output_bias
    output_layer = sigmoid(output_layer_input)

    # Backpropagation
    output_error = y - output_layer
    output_delta = output_error * sigmoid_derivative(output_layer)
    hidden_error = output_delta.dot(output_weights.T)
    hidden_delta = hidden_error * sigmoid_derivative(hidden_layer)

    # Update weights and biases
    output_weights += hidden_layer.T.dot(output_delta) * learning_rate
    output_bias += np.sum(output_delta, axis=0, keepdims=True) * learning_rate
    hidden_weights += X.T.dot(hidden_delta) * learning_rate
    hidden_bias += np.sum(hidden_delta, axis=0, keepdims=True) * learning_rate

# Results
print("Input:")
print(X)
print("\nOutput:")
print(output_layer)
```
