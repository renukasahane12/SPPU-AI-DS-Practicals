```python
import numpy as np

np.random.seed(42)

class XORNetwork:
    def __init__(self):
        self.W1 = np.random.randn(2, 2) * 0.1
        self.b1 = np.zeros((1, 2))
        self.W2 = np.random.randn(2, 1) * 0.1
        self.b2 = np.zeros((1, 1))

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def sigmoid_derivative(self, x):
        return x * (1 - x)

    def forward(self, X):
        self.z1 = np.dot(X, self.W1) + self.b1
        self.a1 = self.sigmoid(self.z1)
        self.z2 = np.dot(self.a1, self.W2) + self.b2
        self.a2 = self.sigmoid(self.z2)
        return self.a2

    def backward(self, X, y, output, lr):
        output_error = y - output
        output_delta = output_error * self.sigmoid_derivative(output)

        z1_error = output_delta.dot(self.W2.T)
        z1_delta = z1_error * self.sigmoid_derivative(self.a1)

        self.W2 += lr * self.a1.T.dot(output_delta)
        self.b2 += lr * np.sum(output_delta, axis=0, keepdims=True)
        self.W1 += lr * X.T.dot(z1_delta)
        self.b1 += lr * np.sum(z1_delta, axis=0, keepdims=True)

    def train(self, X, y, epochs, lr):
        for _ in range(epochs):
            output = self.forward(X)
            self.backward(X, y, output, lr)

    def predict(self, X):
        return self.forward(X)


xor_nn = XORNetwork()

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])

xor_nn.train(X, y, epochs=20000, lr=0.5)

predictions = xor_nn.predict(X)
print(predictions)
```
